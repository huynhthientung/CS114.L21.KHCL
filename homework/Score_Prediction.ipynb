```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read *.csv files
df_1 = pd.read_csv("/content/score_training - score_training.csv")
df_2 = pd.read_csv("/content/wecode.csv")
```

```python
# Set maximum results can be shown
pd.set_option('max_rows', 9999)
pd.set_option('max_colwidth', 999)

df_1_fixed = df_1.replace(r'^\s*$', np.NaN, regex = True) # Replace empty strings with NaN values
df_1_fixed = df_1_fixed[df_1_fixed['CK'].notna()] # Only take not NaN rows
print(df_1_fixed) # Print dt after replaced missing values with NaN
print("\n==> Removed ",len(df_1)-len(df_1_fixed)," NaN lines!")
```

```python
id_correct = [] # Store number of correct assigments of each ID student 
for i in df_1_fixed["id"]:
  id_correct.append(len(df_2[df_2.user_id == i][df_2.is_final == True][df_2.pre_score == 10000]))
```

```python
final_pts = df_1_fixed["CK"] # Get final-term scores column 
```

```python
# Show relationship between final-term scores and number of correct assigments of each student with scatter plot
fig = plt.figure()
plt.scatter(id_correct,final_pts)
plt.title("Scatterplot Relationship between Final Term Scores and Correct Assigments")
plt.xlabel("Correct Assigments")
plt.ylabel("Final Term Scores")
fig.set_figheight(9)
fig.set_figwidth(11)
plt.show()
```
